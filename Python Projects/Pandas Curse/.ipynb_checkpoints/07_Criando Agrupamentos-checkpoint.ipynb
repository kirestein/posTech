# Criando Agrupamentos

```python
import pandas as pd
```

```python
dados = pd.read_csv('Datas/aluguel_residencial.csv', sep=';')
dados.head(10)
```

### https://pandas.pydata.org/pandas-doc/stable/api.html#api-dataframe-stats

```python
round(dados['Valor'].mean(), 2)
```

```python
bairros = dados['Bairro'].unique()
bairros
```

```python
select = dados['Bairro'].isin(bairros)
print(select)
new_dados = dados[select]
print(new_dados)
```

```python
dados['Bairro'].drop_duplicates()
```

```python
grupo_bairro = new_dados.groupby('Bairro')
grupo_bairro.groups
```

```python
[print(f'{bairro} -> {dados.Valor.mean():.2f}\n') for bairro, dados in grupo_bairro]
```

```python
grupo_bairro['Valor'].mean().round(2)
```

## Estatística Descritiva

```python
grupo_bairro['Valor'].describe().round(2).head(10)
```

```python
grupo_bairro['Valor'].aggregate(['min', 'max', 'sum'])
```
