# Fazendo a leitura de outros tipos de arquivos utilizando o pandas

```python
import pandas as pd
```

## Lendo um Json

```python
json = open('dados/aluguel.json')
print(json.read())
```

## var.columns.name = 'nome' serve para dar nome à primeira coluna

```python
df_json = pd.read_json('dados/aluguel.json')
df_json.columns.name = 'Id'
df_json
```

## Lendo arquivo do tipo texto com Pandas

```python
txt = open('dados/aluguel.txt')
txt.read()
```

```python
df_txt = pd.read_table('dados/aluguel.txt')
df_txt.columns.name = 'Id'
df_txt
```

## Lendo arquivo tipo xls com Pandas

```python
xlsx = open('dados/aluguel.xlsx')
xlsx.read(10)
```

```python
df_xlsx = pd.read_excel('dados/aluguel.xlsx')
df_xlsx.columns.name = 'Id'
df_xlsx
```

```python
df_html = pd.read_html('dados/dados_html_1.html')
df_html[0]
```
